```python
import numpy as np
import pandas as pd
import jieba
import jieba.posseg as pseg
from jieba import analyse
```

```python
hotel_comments_ori = pd.read_csv('./translated_agoda_comments_all_city.csv', 
                  sep = ",", encoding = "UTF-8")
```

```python
hotel_comments = hotel_comments_ori[['飯店名稱','hotel_id','國籍', '負評','正評','綜合評論','縣市','鄉鎮','整體評分', '單項評分_整潔度', '單項評分_舒適程度',
       '單項評分_住宿地點', '單項評分_設施', '單項評分_員工素質', '單項評分_性價比']]
```

```python
# 加入自定義字典
jieba.load_userdict("./data/traveldict.txt") #需UTF-8，可以存在此
# 加入自定義停用詞
#jieba.analyse.set_stop_words("./data/stop_words文字雲.txt")
```

```python
hotel_comments['綜合評論'] = hotel_comments['綜合評論'].replace(r'\s+', np.nan, regex=True)
# hotel_comments['正評'] = hotel_comments['正評'].replace(r'\s+', np.nan, regex=True)
# hotel_comments['負評'] = hotel_comments['負評'].replace(r'\s+', np.nan, regex=True)

replacements = ["nil", "none", "nope", "no","","無","沒有","沒有什麼","沒","沒什麼","nothing","沒有任何","不"]
for word in replacements:
    hotel_comments['綜合評論'].replace(word, np.nan, inplace=True)
    # hotel_comments['正評'].replace(word, np.nan, inplace=True)
    # hotel_comments['負評'].replace(word, np.nan, inplace=True)

```

```python
%%time
import re

#正規表達式，只取中文
hotel_comments['綜合評論_cut']= hotel_comments['綜合評論'].str.replace(r'([^\u4e00-\u9fa5]+)','', regex=True)

# 加入stop words
def stopwordlist():
    stopwords = [line.strip() for line in open('./data/stop_words_文字雲.txt',encoding='UTF-8').readlines()]
    return stopwords
stopwords = stopwordlist()

# 斷詞並同時去除stop words
hotel_comments['綜合評論_cut'] = hotel_comments['綜合評論_cut'].astype(str).apply(lambda x : [i for i in jieba.cut(x, cut_all=False) if i not in stopwords])

# 擷取關鍵字TF-IDF (top 15)
hotel_comments['綜合評論key_word'] = hotel_comments['綜合評論_cut'].astype(str).apply(lambda x : [i for i in jieba.analyse.extract_tags(x, topK=15)])
```

```python
hotel_comments['負評_cut']= hotel_comments['負評'].str.replace(r'([^\u4e00-\u9fa5]+)','', regex=True)

# 斷詞並同時去除stop words
hotel_comments['負評_cut'] = hotel_comments['負評_cut'].astype(str).apply(lambda x : [i for i in jieba.cut(x, cut_all=False) if i not in stopwords])

# 擷取關鍵字TF-IDF (top 15)
hotel_comments['負評key_word'] = hotel_comments['負評_cut'].astype(str).apply(lambda x : [i for i in jieba.analyse.extract_tags(x, topK=15)])
```

```python
hotel_comments['正評_cut']= hotel_comments['正評'].str.replace(r'([^\u4e00-\u9fa5]+)','', regex=True)

# 斷詞並同時去除stop words
hotel_comments['正評_cut'] = hotel_comments['正評_cut'].astype(str).apply(lambda x : [i for i in jieba.cut(x, cut_all=False) if i not in stopwords])

# 擷取關鍵字TF-IDF (top 15)
hotel_comments['正評key_word'] = hotel_comments['正評_cut'].astype(str).apply(lambda x : [i for i in jieba.analyse.extract_tags(x, topK=15)])
```

```python
#去除綜合評論_cut的list為空的row
booking_comments = hotel_comments[hotel_comments['綜合評論_cut'].apply(lambda x: len(x) != 0)].reset_index(drop=True)
```

```python
#匯出csv
booking_comments.to_csv('agoda_comments_分詞_文字雲0724.csv', index=False, encoding='utf_8_sig')
```
